```python
import pandas as pd
```

```python
df = pd.read_excel (r"C:\Users\HP\Desktop\boat file.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df["Manufacturer"].str.strip()
```

```python
df["Location"] = df["Location"].str.strip()
```

```python
#Splitting Location into country, region & city
location_split = df["Location"].str.split("»", expand=True)
```

```python
df["Country"] = location_split[0].str.strip()
df["Region"] = location_split[1].str.strip() if len(location_split.columns) > 1 else None
df["City"] = location_split[2].str.strip() if len(location_split.columns) > 2 else None
```

```python
print(df)
```

```python
df["Country"] = df["Country"].str.normalize('NFKD').str.encode('ascii', errors = 'ignore').str.decode('utf-8')
df["Region"] = df["Region"].str.normalize('NFKD').str.encode('ascii', errors = 'ignore').str.decode('utf-8')
df["City"] = df["City"].str.normalize('NFKD').str.encode('ascii', errors = 'ignore').str.decode('utf-8')
```

```python
print(df)
```

```python
df
```

```python
df[['Currency', 'Amount']] = df['Price'].str.split(' ', expand=True)
df
```

```python
#Total amount of boats based on maufacturer
total_amount_per_type_manufacturer = df.groupby(['Boat Type', 'Manufacturer'])['Amount'].sum().reset_index()
print(total_amount_per_type_manufacturer)
```

```python
#Most Used Material
materials_count = df['Material'].value_counts()
print('Materials and Their Count: ')
print(materials_count)
```

```python
#Determine Currency Count so we can convert to USD
Currency_count = df['Currency'].value_counts()
print('Currency and Their Count: ')
print(Currency_count)
```

```python
#Replacing unidentified currency Ã‚Â£ with GBP
df['Currency'] = df ['Currency'].str.replace('Ã‚Â£', 'AUD')
df
```

```python
#Converting all Exchange Rate to USD
df['Amount'] = pd.to_numeric(df['Amount'].str.replace('[^\d.]', ''),errors='coerce')

exchange_rate = {
    'EUR': 1.12,
    'CHF': 1.03,
    'GBP': 1.30,
    'DKK': 0.15
}

df['USD Amount'] = df['Amount'] * df['Currency'].map(exchange_rate)
df
```
